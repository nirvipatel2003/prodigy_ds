```python
total_null=data.isnull().sum().sort_values(ascending=False)
percent=((data.isnull().sum()/data.isnull().count())*100).sort_values(ascending=False)
print("Total Records=",data.shape[0])
missing_data=pd.concat([total_null,percent.round(2)],axis=1,keys=['Total Missing','In Percent'])
missing_data.head(10)
```

```python
data.dropna(subset=['text'],inplace=True)
total_null=((data.isnull().sum()/data.isnull().count())*100).sort_values(ascending=False)
print("Total Records=",data.shape[0])
missing_data=pd.concat([total_null,percent.round(2)],axis=1,keys=['Total Missing','In Percent'])
missing_data.head(10)
```

```python
t0=data[data['sentiment']=="Negative"]
t1=data[data['sentiment']=="Positive"]
t2=data[data['sentiment']=="Irrelevant"]
t3=data[data['sentiment']=="Neutral"]
```

```python
t0.shape,t1.shape,t2.shape,t3.shape
```

```python
t0=t0[:int(t0.shape[0]/12)]
t1=t1[:int(t1.shape[0]/12)]
t2=t2[:int(t2.shape[0]/12)]
t3=t3[:int(t3.shape[0]/12)]
```

```python
t0.shape,t1.shape,t2.shape,t3.shape
```

```python
data=pd.concat([t0,t1,t2,t3],axis=0)
data
```
